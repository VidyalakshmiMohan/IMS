```python
print("Enter the product through Product ID ")
list_items=[]
sales_record={}
total_cost=0
if(1):
    #prod_id=int(input("Enter the Id : "))
    p='y'
    #Until the customer the wishes to order they can order  
    while(p.lower()!='n'):
        prod_id=int(input("Enter the Id : "))
        while( prod_id not in record):
            print("The Id you have entered is not available in the inventory")
            prod_id=int(input("Enter the Id : "))
        qty=int(input("Enter the quantity : "))
        check_qty=record[prod_id]['Quantity']
        if(qty>check_qty):
            print("Specified Quantity is not available ")
            if(check_qty==0):
                print ("The item is not available")
            else:
                
                print("Only",check_qty,"is available")
                #Suggesting the customer to order with less qty
                print("Do you like to order the ",record[prod_id]['Quantity'], "or lesser")
                print("Press y or Y")
                if(input().strip().lower()=='y'):
                    qty=int(input("Enter the quantity : "))
                    record[prod_id]['Quantity']-=qty
                    print("Quantity : ",record[prod_id]['Quantity'])
                    #total price of the product*quantity
                    price_per_product=qty*record[prod_id]['Price']
                    print("Price    :",price_per_product)
                    total_cost+=price_per_product
                    #Storing the items details along with price and qty
    
                    if prod_id in sales_record:
                        sales_record[prod_id]["Quantity"]+=qty 
                        sales_record[prod_id]["Total_Cost"]+=price_per_product
                        
                    else:
                        sales_record[prod_id] = {'Product_Name':record[prod_id]['Product_Name'],'Price':record[prod_id]['Price'],'Quantity':qty,'Total_Cost':price_per_product}
                        
                else:
                    break
        else:
            #When requested quantiy is available
            record[prod_id]['Quantity']-=qty
            print("Quantity : ",record[prod_id]['Quantity'])
            price_per_product=qty*record[prod_id]['Price']
            print("Price    :",price_per_product)
            total_cost+=price_per_product
            if prod_id in sales_record:
                sales_record[prod_id]["Quantity"]+=qty 
                sales_record[prod_id]["Total_Cost"]+=price_per_product
                
            else:
                sales_record[prod_id] = {'Product_Name':record[prod_id]['Product_Name'],'Price':record[prod_id]['Price'],'Quantity':qty,'Total_Cost':price_per_product}

        print("If you don't want to continue ordering Press N or n :")
        p=input().strip()
print("Thanks for Shopping")
```

# *ADDING INTO THE INVENTORY*

```python
yes="yes"

while(yes.lower()=='yes'or yes.lower()=='y'):
    prod_id=int(input("Enter the Id : "))
    qty=int(input("Enter the quantity : "))
    if str(prod_id) in record:
        record[str(prod_id)]["Quantity"]+=qty
    else:
        
        prod_name=input("Enter the Product Name : ")
        price=int(input("Enter the Price : "))
        manufacture_date=input("Enter Manufacturing date in the format DD-MM-YYYY : ").strip()
        expiry_date=input("Enter Expiry date in the format DD-MM-YYYY : ").strip()
        record[prod_id]={'Product_Name':prod_name,'Price':price,'Quantity':qty,'Manufacture_date':manufacture_date,'Expiry_date':expiry_date}
    print("Do you like to continue adding into inventory")
    print("Print Yes or Y to continue")
    yes=input().strip()
```
